# Purpose
This notebook plots the results from the C++ implementation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import the data
float_values = []

with open("edge_energy.txt", "r") as file:
    for line in file:
        float_values.append(float(line.strip()))
```

```python
# edges
plt.plot(float_values)
plt.xlabel("Iteration")
plt.ylabel("Energy")
plt.title("Simulated annealing for edges")
```
