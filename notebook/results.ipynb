# Purpose
This notebook plots the results from the C++ implementation.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import the data
edge_energy, corner_energy = [], []

with open("edge_energy.txt", "r") as file:
    for line in file:
        edge_energy.append(float(line.strip()))

with open("corner_energy.txt", "r") as file:
    for line in file:
        corner_energy.append(float(line.strip()))
```

```python
# edges
plt.plot(edge_energy)
plt.xlabel("Iteration")
plt.ylabel("Energy")
plt.title("Simulated annealing for edges")
```

```python
# corners
plt.plot(corner_energy)
plt.xlabel("Iteration")
plt.ylabel("Energy")
plt.title("Simulated annealing for corners")
```
